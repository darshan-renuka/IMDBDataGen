```python
import pandas as pd
```

```python
# Read 500000 records from MergedData-APr4 and pick 50000 unique movies from the list
i=1
movie_df = pd.DataFrame()
chunk_size = 5
total_req_size = 50
for chunk in pd.read_csv('FinalMergedIMDBMovieData-4Apr.csv', low_memory=False, chunksize=chunk_size):
    print('Chunks left - ', str(total_req_size/i))
    for row, data in chunk.iterrows():
        
        temp_df = pd.DataFrame(data)
        print(temp_df)
        print('Size of chunk - ', str(temp_df.shape))
        if i==0:
            movie_df = temp_df
        else:
            movie_df = pd.concat([movie_df, temp_df])
            
        if i > total_req_size:
            break
    print(i)
    i+=chunk_size
    
print(movie_df.shape)
print(movie_df.head())
```
